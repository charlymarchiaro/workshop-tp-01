## Expression

```python
from sympy import *
```

```python
x=symbols('x')
y=x**2+3*x**2-4
plot(y)
```

```python
simplify(y)
```

```python
solve(y,x)
```

```python
integrate(y)
```

```python
diff(y)
```

```python
y=sin(x)
plot(y)
```

```python
y=exp(x)
plot(y)
```
